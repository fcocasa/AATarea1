## 2.1 Juego
Las damas es un juego para dos personas en un tablero de 64 casillas de 8×8 celdas (el mismo que se utiliza para jugar al ajedrez). El tablero se coloca de manera que cada jugador tenga una casilla blanca en su parte inferior derecha.

Cada jugador dispone de 12 piezas de un mismo color (unas blancas y las otras negras) que al principio de la partida se colocan en las casillas negras de las tres filas más próximas a él. El objetivo del juego de damas es capturar las fichas del oponente o acorralarlas para que los únicos movimientos que puedan realizar sean los que lleven a su captura.

Se considera una versión simplificada del juego:
Se juega por turnos alternos. Empieza a jugar quien tiene las fichas claras (blancas). En su turno cada jugador mueve una pieza propia, y debe mover a menos que no tenga ningún movimiento posible.
Todos los movimientos son en diagonal. Cada peón puede mover una casilla en diagonal hacia adelante, nunca hacia atrás, se realiza un movimiento por turno. No se puede ocupar la casilla ocupada por otra pieza.

Cuando un peón propio tiene (en diagonal)  como vecina a una figura  contraria e inmediatamente  detrás de ella hay una casilla vacía se la come  saltando por encima de ella hacia la casilla vacía inmediatamente detrás. Los peones  sólo pueden saltar en diagonal, hacia adelante, no hacia atrás. La figura contraria saltada se retira del tablero en el mismo turno del salto, después del cual pasa el turno al contrario. No hay obligación de comer.

El juego termina cuando ninguno de los jugadores puede realizar un movimiento, y el ganador es aquel jugador que termina con más fichas en el tablero.



## 2.2 Tablero
El tablero se representa como una matriz de 8x8 donde ‘b’ representa una ficha blanca, ‘n’ representa una ficha negra y ‘-’ representa un espacio vacío en el tablero. Se consideran atributos relevantes del tablero tanto la cantidad de fichas de cada color (fichasB y fichasN) como la cantidad de amenazas a fichas de cada color (amenazadasB y amenazadasN). Una ficha está amenazada si existe un movimiento posible del contrario en el tablero actual que implica comer esta ficha. Si una ficha es amenazada por dos fichas contrarias se cuenta como dos amenazas.

## 2.4 Algoritmo
La función de valoración para la inteligencia artificial se representa como la suma de cada uno de los atributos del tablero multiplicados por diferentes pesos wi, los cuales a la hora de entrenar a la IA son inicialmente aleatorios y se van actualizando según se realizan jugadas.

```python
Vop(tablero) = w0 * fichasB + w1 * fichasN + w2 * amenazadasB + w3 * amenazadasN
```

Al finalizar cada turno, es decir, al haber hecho un movimiento el jugador y otro el contrario, se ajustan los pesos tomando en cuenta el valor del tablero del cuál se partió y el valor del tablero que fue devuelto en el siguiente turno, de manera que el resultado del primero, con los nuevos valores de los pesos, se aproxime al valor del segundo. 

Para que la función de valoración se mantenga en el rango [-1,1] (valores que se devuelven si una partida fue perdida o ganada respectivamente), al recalcular estos pesos, se dividen entre la norma de su vector.
El ajuste a los pesos se realiza mediante el algoritmo de mínimos cuadrados:

```python
Para cada <t, Vent(t)>
    Calculamos Vop(t) con los pesos actuales
    Para cada peso wi
        wi ← wi + μ(Vent(t)-Vop(t)) xi 
    Fin Para
    Para cada peso wi
        wi ← wi / ||w||
Fin Para
```
Donde:

μ es el factor de aprendizaje, es decir a mayor μ mayor es el ajuste

Vent(t) es el valor del tablero antes de realizar el movimiento

Vop(t) es el valor del tablero luego de que el contrario realiza su movimiento en reacción al elegido por el jugador

x es el vector de valores de los atributos del tablero correspondiente a Vop(t)

Para las pruebas se utilizaron diferentes valores de μ, así como diferentes maneras de implementación de la tasa de enfriamiento (por ejemplo, decrecer el valor de μ después de n partidas, después de n partidas ganadas consecutivas)

Para seleccionar el movimiento a realizar, se hace uso del algoritmo minimax de dos niveles, donde para todos los movimientos posibles MP del jugador se calculan los valores de los tableros de todos los movimientos posibles del oponente dado que ya fue realizado el movimiento MP, utilizando la función de valoración definida anteriormente. Si el oponente jugara de manera óptima, para cada movimiento MP, realizaría el movimiento que minimiza el valor devuelto por la función de valoración (ya que sería el peor tablero para el jugador). Entonces, una vez obtenidos todos los mínimos devueltos por el oponente, el jugador deberá elegir el movimiento MPopt tal que maximice el valor del conjunto de valores que le devolvió su oponente. Este movimiento MPopt es el que efectivamente se realiza. 

## 2.4 Contrarios
Se implementaron dos principales jugadores contrarios: Un jugador que realiza sus movimientos de forma aleatoria y uno que utiliza la misma estrategia que la IA, solo que con pesos fijos, que se actualizan a los de la IA (cambiando de signo sus valores, ya que es el jugador contrario) cada una cierta cantidad de partidas.

El jugador aleatorio implementado, cada vez que es su turno de jugar, encuentra todos sus posibles movimientos a realizar, guardándolos en un lista. A continuación, se utiliza la función random.choice() para, de todos esos movimientos posibles, elegir uno al azar. El jugador procede a realizar el movimiento elegido.

Para el jugador que copia sus valores de la IA, la elección del movimiento se basa en el cual devuelve un valor del tablero calculado siguiendo la estrategia de minimax de 2 niveles explicada en la Sección 2.3.

En la primera subsección se expone el funcionamiento del entrenamiento con un contrincante aleatorio, mostrando las diferencias entre los distintos valores de μ, tanto manteniendo el valor fijo o con una cierta tasa de envejecimiento. En la segunda subsección se muestra el comportamiento al entrenar contra una versión anterior del jugador, con las mismas variantes. En las primeras tres subsecciones se entrena el jugador implementado por 250 partidas, usando fichas blancas. Se randomizan los pesos iniciales, que quedan fijos para todos los entrenamientos de manera que la comparación sea justa y la comparación se centre en el factor de aprendizaje y la tasa de envejecimiento. Se grafican los resultados de las siguientes situaciones:

- 10 partidas jugadas ⇾ μ/10:
El factor de aprendizaje cuenta con una tasa de envejecimiento independiente del rendimiento, es decir que cada 10 partidas jugadas, ya sean ganadas o perdidas, se actualiza el valor de μ dividiendolo entre diez.
- μ=0.1 constante:
El factor de aprendizaje no tiene tasa de envejecimiento, por lo tanto se matiene en 0.1 a lo largo de todas las partidas.
- 4 partidas seguidas ganadas ⇾ μ/4:
La tasa de envejecimiento de μ responde al comportamiento del jugador, dividiendo el factor de aprendizaje entre cuatro cada vez que ocurren cuatro victorias consecutivas.

### 3.3. Jugador vs Valores Fijos: Nueva evaluación de tablero final

Como ya fue explicado, este algoritmo implementa un cambio al algoritmo anterior en la manera de dar valor a los tableros finales. En lugar de devolver los valores -1, 0 o 1 cuando se pierde, empata o gana una partida respectivamente, se devuelve la diferencia de fichas de los distintos jugadores.

Además, se incluyen nuevas implementaciones de los algoritmos en cuanto al factor de aprendizaje para buscar mayor eficiencia del jugador:

- 50 partidas jugadas ⇾ μ/10:
La tasa de envejecimiento se enlentece, cambiando solo cada 50 partidas jugadas.
- 50 partidas jugadas ⇾ (μ/10, actualización de pesos):
Además del cambio anterior, se enlentece también la actualización de los pesos del contrario, ocurriendo también cada 50 partidas jugadas.

## 4. Conclusión

Durante los distintos experimentos pudimos observar que hay distintos factores que influyen en el rendimiento del algoritmo. 
Uno de estos es el factor de envejecimiento, el cual influye en la velocidad con la que aprende el jugador, al usarlo en casos donde el oponente mantiene una misma estrategia nos es útil, ya que permite ajustar los pesos de forma más precisa. En el caso donde el jugador se enfrentaba a una versión anterior de sí mismo que se actualizaba cada N partidas descubrimos que el factor de envejecimiento puede ser contraproducente si no es usado de forma correcta, ya que el oponente también mejora, si el factor de aprendizaje se vuelve más pequeño sin tener en consideración que el oponente se actualiza, llega un punto donde el jugador aprende tan poco que no logra ganarle a su oponente.

Otros de estos factores importantes son los valores iniciales, ya que mientras mejores sean estos, más sencillo va a ser que el jugador empiece a ganar. Nos referimos a que los valores son buenos cuando le da valoraciones convenientes para nuestro jugador respecto a los parámetros considerados en la función de valuación, es decir, si el jugador juega con las blancas serían buenos valores si se empieza con una alta valoración para las cantidad de fichas blancas en comparación con el valor de las negras.

Por último se tiene la cantidad de iteraciones, mientras más grande sea el número de iteraciones contra un oponente más ajusten se le realizan a los pesos.

A pesar de que todos estos valores influyen en el comportamiento del algoritmo no son independientes, ya que el que tanto influyen en el algoritmo depende del valor de los otros, por ejemplo, si se tiene un factor de aprendizaje alto en las primera iteraciones permite que el jugador se ajuste más rápidos, pero una vez que se llevan muchas iteraciones si el factor de aprendizaje es muy alto puede generar oscilaciones en el valor de los pesos lo cual puede empeorarlo muy rápido.

También se puede observar con los valores iniciales, si estos valores son malos, al tener un factor de aprendizaje alto lograran que el jugador consiga empezar a ganar partidas en menos iteraciones de las que necesitaría si el factor de aprendizaje fuera menor.

Sin embargo, para mejorar el rendimiento general del algoritmo podría mejorarse volviéndolo más complejo, al agregar nuevos factores al cálculo de evaluación de tableros. Algunos atributos candidatos a agregar son la cantidad de fichas en posiciones centrales del tablero, ya que tienen ventaja en cuanto a movilidad, la cantidad de fichas en el extremo contrario, que ya no pueden moverse por el modelo de juego elegido, o simplemente la cantidad de fichas con y sin posibilidades de movimiento.
